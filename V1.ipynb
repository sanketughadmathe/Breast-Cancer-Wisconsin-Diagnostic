# EDA

```python
# Checking the count of target
sns.countplot(df['diagnosis'])
plt.show()
```

```python
# Checking the count of target
df['diagnosis'].value_counts()
```

```python
# Converting M & B to 0 & 1 respectively
df['diagnosis']= df['diagnosis'].map(dict(M=0, B=1))
```

```python
# Droping Id & Unnamed column
df.drop(['id', 'Unnamed: 32'], axis= 1, inplace= True)
```

```python
# Checking the Correlation
df.corr()
```

```python
# Plotting the Correlation
df.corr()['diagnosis'][2:].sort_values().plot(kind= 'bar', figsize= (18,9))
```

```python
# Plotting the heatmap
plt.figure(figsize= (15,12))
sns.heatmap(df.corr(), cmap= 'coolwarm_r')
```
