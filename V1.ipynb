# Defining X & y

```python
X= df.drop('diagnosis', axis= 1).values
```

```python
y= df['diagnosis'].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.25, random_state= 101)
```

# Scaling the data

```python
from sklearn.preprocessing import MinMaxScaler
scaler= MinMaxScaler()
```

```python
X_train= scaler.fit_transform(X_train)
X_test= scaler.transform(X_test)
```

# ANN

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
```

```python
model= Sequential()

# Input Layer
model.add(Dense(30, activation= 'relu'))

# Hidden Layer
model.add(Dense(15, activation= 'relu'))

# Output Layer
model.add(Dense(1, activation= 'sigmoid'))

model.compile(loss= 'binary_crossentropy', optimizer= 'adam')
```

```python
model.fit(x=X_train, 
          y=y_train, 
          epochs=600,
          validation_data=(X_test, y_test)
          )
```

```python
model_loss= pd.DataFrame(model.history.history)
model_loss.plot(figsize= (10,7))
```

## As we can see this is a perfect example of Overfiting.

* Key characteristics of Overfitting
    * In the beginning during first couple of epochs, both validation and traning loss is decreasing, that's good which means we haven't yet Overfit to our traning data.
    * As we go along we are decreasing loss in both validdation set and training set.
    * However at a certein point in time notice that our traning loss in blue is still going down.
    * But our validation loss is beginning to increase.
    * That basically tells us that we area Overfitting to our traning dataset.
    * So we haveclear indication that we are traning for too many epochs.
    * Notice that validation loss gtes worse and worse after these epochs.<br><br>
* So what we are going do is we're going to see if we can use early stopping.
* We'll use TensorFlow keras callback to actually based on your validation loss stop the traning before it gets out of hand as shown above

```python
model= Sequential()

# Input Layer
model.add(Dense(30, activation= 'relu'))

# Hidden Layer
model.add(Dense(15, activation= 'relu'))

# Output Layer
model.add(Dense(1, activation= 'sigmoid'))

model.compile(loss= 'binary_crossentropy', optimizer= 'adam')
```

```python
from tensorflow.keras.callbacks import EarlyStopping
```

```python
early_stop = EarlyStopping(monitor= 'val_loss', mode= 'min', verbose= 1, patience= 25)
```

```python
model.fit(x=X_train,
          y=y_train,
          epochs=600,
          validation_data=(X_test, y_test), verbose=1,
          callbacks=[early_stop]
          )
```

```python
model.history.history
```

```python
model_loss= pd.DataFrame(model.history.history)
model_loss.plot(figsize= (10,7))
```

## Other method to stop Overfitting
* Other tiing we can do to try to help prevent Over fitting is add a dropout layers.
* So a dropout layers will turn off a percentage of neurons randomly.

```python
from tensorflow.keras.layers import Dropout
```

```python
model= Sequential()

# Input Layer
model.add(Dense(30, activation= 'relu'))
# Dropout Layer
model.add(Dropout(0.5))

# Hidden Layer
model.add(Dense(15, activation= 'relu'))
# Dropout Layer
model.add(Dropout(0.5))

# Output Layer
model.add(Dense(1, activation= 'sigmoid'))

model.compile(loss= 'binary_crossentropy', optimizer= 'adam')
```

```python
model.fit(x=X_train,
          y=y_train,
          epochs=600,
          validation_data=(X_test, y_test), verbose=1,
          callbacks=[early_stop]
          )
```

```python
model.history.history
```

```python
model_loss= pd.DataFrame(model.history.history)
model_loss.plot(figsize= (10,7))
```

* Note that the traning loss and the validation loss are both quickly going down and  they are essentially flattening out at the same rate.
* This is exactly the kind of behaviour we want to see
* This is much improved over what we saw earlier.<br><br>
* So note that adding in the early stopping and adding those dropout layers has significantly increased performance.
